Using merge statements

Insert any new records being received


Update any existing records for which new records have been received


And delete any non needed rows all in one statement

```
drivers_day1_df = spark.read \
.option("inferSchema", True) \
.json("/mnt/formula1brenden/raw/2021-03-28/drivers.json")\
.filter("driverId <= 10")\
.select("driverId", "dob", col("name.forename").alias("forename"), col("name.surname").alias("surname"))
```

```
display(drivers_day1_df )
```

```
from pyspark.sql.functions import *

drivers_day2_df = spark.read \
.option("inferSchema", True) \
.json("/mnt/formula1brenden/raw/2021-03-28/drivers.json")\
.filter("driverId BETWEEN 6 AND 25")\
.select("driverId", "dob", upper("name.forename").alias("forename"), upper("name.surname").alias("surname"))
```

```
display(drivers_day2_df)
```

```
drivers_day3_df = spark.read \
.option("inferSchema", True) \
.json("/mnt/formula1brenden/raw/2021-03-28/drivers.json")\
.filter("driverId BETWEEN 1 AND 5 or driverId BETWEEN 16 and 20")\
.select("driverId", "dob", upper("name.forename").alias("forename"), upper("name.surname").alias("surname"))
```

```
display(drivers_day3_df)
```

```
drivers_day1_df.createOrReplaceTempView("drivers_day1")
```

```
%sql select * from drivers_day1
```

```
drivers_day2_df.createOrReplaceTempView("drivers_day2")
```

```
%sql 
create table if not exists f1_demo.drivers_merge(
driverId int,
dob date,
forename string,
surname string,
createdDate date,
updateddate date
)
using delta
```

Day 1

```
%sql 
MERGE INTO f1_demo.drivers_merge tgt
USING drivers_day1 updates
ON tgt.driverId = updates.driverId
when matched then
  update set tgt.dob = updates.dob,
              tgt.forename = updates.forename,
              tgt.surname = updates.surname,
              tgt.updateddate = current_timestamp
WHEN NOT MATCHED
  THEN INSERT (driverId, dob, forename, surname, createdDate) Values (driverId, dob, forename, surname, current_timestamp)
```

```
%sql 
select * from f1_demo.drivers_merge
```

Day 2

```
%sql 
MERGE INTO f1_demo.drivers_merge tgt
USING drivers_day2 updates
ON tgt.driverId = updates.driverId
when matched then
  update set tgt.dob = updates.dob,
              tgt.forename = updates.forename,
              tgt.surname = updates.surname,
              tgt.updateddate = current_timestamp
WHEN NOT MATCHED
  THEN INSERT (driverId, dob, forename, surname, createdDate) Values (driverId, dob, forename, surname, current_timestamp)
```

```
%sql 
select * from f1_demo.drivers_merge
```

```
from delta.tables import *

deltaTable = DeltaTable.forPath(spark, '/mnt/formula1brenden/demo/drivers_merge')

deltaTable.alias("tgt").merge(
    drivers_day3_df.alias('updates'),
    'tgt.driverId = updates.driverId'
  ) \
  .whenMatchedUpdate(set =
    {
      "dob" : "updates.dob",
              "forename": "updates.forename",
              "surname": "updates.surname",
              "updatedDate": "current_timestamp"
    }
  ) \
  .whenNotMatchedInsert(values =
    {
      "dob" : "updates.dob",
              "forename": "updates.forename",
              "surname": "updates.surname",
              "createdDate": "current_timestamp"
    }
  ) \
  .execute()
```
