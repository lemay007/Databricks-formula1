```
from pyspark.sql.types import *

constructors_schema = "constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRIng"

constructor_df = spark.read \
.schema(constructors_schema) \
.json(f"{raw_folder_path}/constructors.json")
```

#### Step 2 - Drop unwanted columns from the dataframe

```
constructor_droped_df = constructor_df.drop("url")
```

#### Step 3 - Rename columns and ingest data

```
from pyspark.sql.functions import *

constructor_final_df = add_ingestion_date(constructor_df).withColumnRenamed("constructorId","constructor_id")\
.withColumnRenamed("constructorRef", "constructor_ref")
```

#### Final Step - Write data to datalake

```
constructor_final_df.write.mode("overwrite").parquet(f"{processed_folder_path}/constructors")
```
