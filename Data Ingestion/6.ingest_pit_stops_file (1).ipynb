```
#Let's import some packages
from pyspark.sql.types import *
from pyspark.sql.functions import *

pit_stop_schema = StructType(fields = [
    StructField("raceId", IntegerType(), False),
    StructField("driverId", IntegerType(), True),
    StructField("stop", StringType(), True),
    StructField("lap", IntegerType(), True),
    StructField("duration", StringType(), True),
    StructField("milliseconds", IntegerType(), True)
])

pit_stops_df = spark.read \
.schema(pit_stop_schema) \
.option("multiLine", True) \
.json(f"{raw_folder_path}/pit_stops.json")
```

```
display(pit_stops_df)
```

#### Step 2 - Rename columns and add ingestion_date

```
final_df = add_ingestion_date(pit_stops_df).withColumnRenamed("driverId", "driver_id") \
.withColumnRenamed("raceId", "race_id") 
```

#### Step 3 - Write data to datalake

```
final_df.write.mode("overwrite").parquet(f"{processed_folder_path}/pit_stops")
```

```
final_df.write.mode("overwrite").format("parquet").saveAsTable("f1_processed.pit_stopsSQL")
```
